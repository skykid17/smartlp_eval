## Retrieval Evaluation Analysis Notebook

### Import Required Libraries


```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import scipy.stats as stats
from scipy.stats import ttest_rel
```

```python
# Notebook is in notebooks/ so go up one level to project root
BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
INPUT_DIR = BASE_DIR / "data/eval/input"
OUTPUT_DIR = BASE_DIR / "data/eval/output"

files = {
    "text": OUTPUT_DIR / "retriever_results_text.json",
    "vector": OUTPUT_DIR / "retriever_results_vector.json",
    "hybrid": OUTPUT_DIR / "retriever_results_hybrid.json",
}

results = {}

for method, path in files.items():
    with open(path, "r") as f:
        results[method] = json.load(f)

print("Loaded datasets:", {m: len(v) for m,v in results.items()})
```

#### Convert to Dataframe

```python
rows = []

for method, entries in results.items():
    for entry in entries:
        rows.append({
            "method": method,
            "input": entry["input"],
            "precision": entry["contextual_precision"]["score"],
            "recall": entry["contextual_recall"]["score"],
            "relevancy": entry["contextual_relevancy"]["score"],
        })

df = pd.DataFrame(rows)

# Set categorical order for consistent display
df["method"] = pd.Categorical(df["method"], categories=["text", "vector", "hybrid"], ordered=True)

df.head()
```

#### Basic Sanity Check

```python
print("Total rows:", len(df))
print(df.groupby("method").size())

df.describe()
```

#### Mean metrics per retriever

```python
summary = df.groupby("method", observed=False)[["precision","recall","relevancy"]].mean()
summary = summary.reindex(["text", "vector", "hybrid"])
summary
```

```python
summary.plot(kind="bar")
plt.ylim(0,1)
plt.title("Retriever Quality Comparison (DeepEval Metrics)")
plt.ylabel("Mean Score")
plt.show()
```

#### Precision vs Recall Scatter

```python
sns.scatterplot(
    data=df,
    x="recall",
    y="precision",
    hue="method",
    hue_order=["text", "vector", "hybrid"],
    alpha=0.7
)

plt.title("Precision vs Recall Tradeoff")
plt.xlim(0,1)
plt.ylim(0,1)
plt.show()
```

#### Distribution plots per metric

```python
for metric in ["precision","recall","relevancy"]:
    sns.kdeplot(data=df, x=metric, hue="method", hue_order=["text", "vector", "hybrid"], fill=True)
    plt.title(f"Distribution of {metric} by Retriever")
    plt.xlim(0,1)
    plt.show()
```

#### Statistical significance test (ANOVA) 
Tells us if the the mean scores of at least one retriever significantly different from the others

```python
for metric in ["precision","recall","relevancy"]:
    groups = [df[df.method==m][metric] for m in ["text", "vector", "hybrid"]]
    f,p = stats.f_oneway(*groups)
    print(metric.upper(), "p-value =", round(p,6))
```

There is an extremely strong statistical difference in performance between the retrievers for all three metrics.

#### Win-Rate Analysis


```python
winners = []

for _, group in df.groupby("input"):
    best = group.sort_values("relevancy", ascending=False).iloc[0]
    winners.append(best["method"])

winrate = pd.Series(winners, dtype=pd.CategoricalDtype(categories=["text", "vector", "hybrid"], ordered=True)).value_counts(normalize=True)
winrate = winrate.reindex(["text", "vector", "hybrid"])
winrate
```

#### Metric Correlation Analysis

```python
corr = df[["precision","recall","relevancy"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Metric Correlation Matrix")
plt.show()
```

###  Does adding BM25 text search to vector retrieval (Hybrid = Text + Vector) meaningfully improve retrieval quality over Vector alone?

```python
df_v = df[df.method=="vector"].copy()
df_h = df[df.method=="hybrid"].copy()

# ensure alignment by query text
merged = df_v.merge(
    df_h,
    on="input",
    suffixes=("_vector","_hybrid")
)

len(merged)
```

```python
for metric in ["precision","recall","relevancy"]:
    merged[f"delta_{metric}"] = merged[f"{metric}_hybrid"] - merged[f"{metric}_vector"]

merged[[c for c in merged.columns if "delta" in c]].describe()
```

```python
for metric in ["precision","recall","relevancy"]:
    sns.histplot(merged[f"delta_{metric}"], kde=True)
    plt.axvline(0, color="red", linestyle="--")
    plt.title(f"Hybrid − Vector delta: {metric}")
    plt.show()
```

```python
wins = {}

for metric in ["precision","recall","relevancy"]:
    wins[metric] = {
        "hybrid_better": (merged[f"delta_{metric}"] > 0).mean(),
        "equal": (merged[f"delta_{metric}"] == 0).mean(),
        "vector_better": (merged[f"delta_{metric}"] < 0).mean(),
    }

pd.DataFrame(wins)
```

Hybrid wins.

```python
for metric in ["precision","recall","relevancy"]:
    t,p = ttest_rel(
        merged[f"{metric}_hybrid"],
        merged[f"{metric}_vector"]
    )
    print(metric.upper(), "p-value =", round(p,10))
```

Improvement in using hybrid is statistically significant.
